###¿Existe relacion lienal entre estas varaibles ?

Para determinar la existencia de una relacion lineal entre las varibles $x$ y $y$ se usa el coeficiente de correlacion de Pearson

$$\rho = r =  \frac{Sxy}{SxSy}=\frac{\sum(x_i-\bar{x})(y_i-\bar{y})}{\sqrt{\sum(x_i-\bar{x})^2}\sqrt{\sum(y_i-\bar{y})^2}}$$

$r→1$  la relacion entre $x$ y $y$ es lineal positiva (directa) **"si $x$ crece $y$  crece"** ($r>0.75$)

$r→-1$ la relacion entre $x$ y $y$ es lineal negativa (inversa) **"si $x$ crece $y$  disminuye"**  ($r<-0.75$) 

$r→0$ No existe relacion linela entre $x,y$

# Inferencia sobre los parametros del modelo de regresion

## Inferencias sobre los coeficientes $\beta_0$ y $\beta_1$

$x,y→ $ variables

image.png

$y=\beta_0+\beta_1x$

image.png

**Objetivo:** ajustar el modelo lineal a los datos "lo mejor posible"

$$E(y)=\beta_0+\beta_1x $$

$E:$ valor esparado

esto se traduce $\hat{y}=\hat{\beta_0}+\hat{\beta_1}x$

$\hat{\beta_0}, \hat{\beta_1}$ son estimaciones de $\beta_0, \beta_0$

Considere el caso en que se hacen $n$ observaciones 

$x=(x_1,x_2,...,x_n), \quad y= (y_1,y_2,y_n)$

para cada observacion

$\hat{y}=\hat{\beta_0}+\hat{\beta_1}x$ es una estimacion $y_i$ el error de la estimacion esta dado por

$$\varepsilon=y_1-\hat{y}_i $$

la suma de los cuadrados del error, esta dada por:

$$SSE=\sum_{i=1}^{n}\varepsilon_i^2 =\sum_{i=1}^{n} (y_1-\hat{y}_i)^2$$

Suma del modelo lineal 

$E(\varepsilon)=0, \quad \text{Var}(\varepsilon)=\sigma^2$ (constante)

$E:$ valor espaerado (media)

$\text{Var}:$ varianza

Bajo estos supuesto, se puede demostrar que:

$$E(\hat{\beta_1}) = \beta_1, \quad \text{Var}(\hat{\beta_1})=\frac{\sigma}{S_{xx}}$$

$$E(\hat{\beta_0})=\beta_0. \quad \text{Var}(\hat{\beta_0})=\frac{\sigma^2\sum_{x_i^2}}{nS_{xx}} $$

$$\text{Cov}(\hat{\beta_0},\hat{\beta_1})=\frac{-\bar{x}\sigma^2}{S_{xx}} $$

En general $\sigma^2$ es desconocido

Se puede demostrar que un estimador de $\sigma^2$ es:

$$S^2=\frac{\sum_{i=1}^{n} (y_1-\hat{y}_i)^2}{n-2}=\frac{SSE}{n-2}$$

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([2, 2, 3, 4, 1, 5 ])
y = np.array([95.75, 110.19, 118.33, 150.92, 85.86, 180.62]) 
xy = np.array([x,y])

def regresion_lineal(x,y):
    media_x=np.mean(x)
    media_y = np.mean(y)
    dx = x-media_x
    dy = y-media_y
    dx2=dx**2
    dy2=dy**2
    r=sum(dx*dy)/(np.sqrt(sum(dx2))*np.sqrt(sum(dy2)))
    b1 = sum(dx*dy)/sum(dx2)
    b0 = media_y-b1*media_x
    x_1 = np.array([np.min(x),np.max(x)])
    y_1 = b0+b1*x_1
    s1=sum(dx2)/len(x)
    s2=sum(dy2)/len(y)
    y_hat=b0+b1*x
    SSE = sum((y-y_hat)**2)
    C00 = sum(x**2)/(5*sum(dx2))
#varianza de la estimacion de b1
    C11 = 1/(sum(dx2))
#Covarianza entre las estimaciones 
    C01 = media_x/(sum(dx2))
    plt.plot(x ,y , "o")
    plt.plot(x_1 ,y_1 , "--")
    plt.grid()
    plt.show()
    print('s1=',s1,), print('s2=',s2),print('r=',r), print('y=',b0,'+',b1,'*x'), print('SSE=',SSE), print('C01 =', C01),print('C11 =',C11),print('C00 =', C00), print('dx2=', dx2), print('media_x=',media_x)


```
